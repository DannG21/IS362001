```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
cumulative_miles = pd.Series([55, 120, 180, 240, 310, 385])
```

```python
daily_miles = cumulative_miles.diff().fillna(cumulative_miles[0])
```

```python
miles_df = pd.DataFrame({
    'Cumulative Miles': cumulative_miles,
    'Daily Miles': daily_miles
})
```

```python
plt.figure(figsize=(10, 6))
plt.plot(miles_df.index, miles_df['Cumulative Miles'], label='Cumulative Miles', marker='o', linestyle='-', color='b')
plt.bar(miles_df.index, miles_df['Daily Miles'], label='Daily Miles', color='red', alpha=0.7)

plt.title('My Bicycle Ride Progress from New York to Key West', fontsize=16)
plt.xlabel('Days', fontsize=12)
plt.ylabel('Miles', fontsize=12)
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
