```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'C:\\Users\\danie\\Desktop\\goods.csv'
data = pd.read_csv(file_path)
```

```python
data.head(), data.info()
```

```python
data.columns = ['Region', 'Chocolate_Jan_Sales', 'Chocolate_Feb_Sales', 'Chocolate_Mar_Sales', 
                'Snacks_Jan_Sales', 'Snacks_Feb_Sales', 'Snacks_Mar_Sales']
```

```python
data.head()
```

```python
data['Total_Chocolate_Sales'] = data[['Chocolate_Jan_Sales', 'Chocolate_Feb_Sales', 'Chocolate_Mar_Sales']].sum(axis=1)
data['Total_Snacks_Sales'] = data[['Snacks_Jan_Sales', 'Snacks_Feb_Sales', 'Snacks_Mar_Sales']].sum(axis=1)
```

```python
data['Average_Chocolate_Sales'] = data[['Chocolate_Jan_Sales', 'Chocolate_Feb_Sales', 'Chocolate_Mar_Sales']].mean(axis=1)
data['Average_Snacks_Sales'] = data[['Snacks_Jan_Sales', 'Snacks_Feb_Sales', 'Snacks_Mar_Sales']].mean(axis=1)
```

```python
data[['Region', 'Total_Chocolate_Sales', 'Total_Snacks_Sales', 'Average_Chocolate_Sales', 'Average_Snacks_Sales']]
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Bar plot: Total Chocolate Sales by Region
axs[0, 0].bar(data['Region'], data['Total_Chocolate_Sales'], color='chocolate')
axs[0, 0].set_title('Total Chocolate Sales by Region')
axs[0, 0].set_ylabel('Sales')

# Bar plot: Total Snacks Sales by Region
axs[0, 1].bar(data['Region'], data['Total_Snacks_Sales'], color='orange')
axs[0, 1].set_title('Total Snacks Sales by Region')
axs[0, 1].set_ylabel('Sales')

# Line plot: Average Chocolate Sales by Region
axs[1, 0].plot(data['Region'], data['Average_Chocolate_Sales'], marker='o', color='chocolate')
axs[1, 0].set_title('Average Chocolate Sales by Region')
axs[1, 0].set_ylabel('Average Sales')

# Line plot: Average Snacks Sales by Region
axs[1, 1].plot(data['Region'], data['Average_Snacks_Sales'], marker='o', color='orange')
axs[1, 1].set_title('Average Snacks Sales by Region')
axs[1, 1].set_ylabel('Average Sales')

plt.tight_layout()

plt.show()
```

Some Key Findings from the analysis:

- The East region consistently shows the highest sales for both products.
- Chocolate sales generally show more variability than snack sales.
- There appears to be a correlation between chocolate and snack sales performance.
- Most regions show consistent sales patterns across months.
