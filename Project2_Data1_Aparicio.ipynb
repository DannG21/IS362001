```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df = pd.read_csv('C:\\Users\\danie\\Downloads\\dummy_devices.csv')
```

```python
# Display basic information about the dataset
print(df.info())
print("\nFirst few rows of the dataset:")
print(df.head())
```

```python
# Clean and preprocess the data
df['Registration Time'] = pd.to_datetime(df['Registration Time'])
```

```python
# Handle the MDM Compliance column
df['MDM Compliance'] = df['MDM Compliance'].fillna(False)
df['MDM Compliance'] = df['MDM Compliance'].astype(bool)
```

```python
# Analysis 1: Count of devices by type
device_types = df['Device Type'].value_counts()
print("\nDevice Types:")
print(device_types)
```

```python
# Analysis 2: Operating System distribution
os_distribution = df['Operating Systems'].value_counts()
print("\nOperating System Distribution:")
print(os_distribution)
```

```python
# Analysis 3: MDM Compliance
mdm_compliance = df['MDM Compliance'].value_counts(normalize=True) * 100
print("\nMDM Compliance Percentage:")
print(mdm_compliance)
```

```python
# Analysis 4: Devices registered per month
df['Registration Month'] = df['Registration Time'].dt.to_period('M')
monthly_registrations = df['Registration Month'].value_counts().sort_index()
print("\nMonthly Device Registrations:")
print(monthly_registrations)
```

```python
# Analysis 5: Most common device models
device_models = df['Device Model'].value_counts().head(10)
print("\nTop 10 Device Models:")
print(device_models)
```

```python
# Visualization: Device Types
plt.figure(figsize=(10, 6))
device_types.plot(kind='bar')
plt.title('Device Types Distribution')
plt.xlabel('Device Type')
plt.ylabel('Count')
plt.tight_layout()
```

```python
# Visualization: Operating System Distribution
plt.figure(figsize=(12, 6))
os_distribution.plot(kind='bar')
plt.title('Operating System Distribution')
plt.xlabel('Operating System')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
```
